# Olid-BR

```python
df_olidbr = pd.read_csv("/content/drive/MyDrive/IC/data/raw/olid-br.csv")
```

```python
df_olidbr.head()
```

```python
df_olidbr.drop(columns=['id', 'is_targeted', 'targeted_type','toxic_spans'], inplace=True)
```

```python
colunms_to_map = df_olidbr.columns[1:]
```

```python
label_map = {"OFF": 1, True: 1,
             "NOT": 0, False: 0}
```

```python
df_olidbr[colunms_to_map] = df_olidbr[colunms_to_map].apply(lambda x: x.map(label_map))
```

```python
df_olidbr.head()
```

```python
words_to_replace = ["USER", "HASHTAG", "URL", "XXXXX"]
```

```python
df_olidbr["text"] = df_olidbr["text"].replace(words_to_replace, " ", regex=True)
```

```python
df_olidbr["text"] = df_olidbr["text"].apply(prep.process_tweet)
```

```python
df_olidbr.dropna(inplace=True)
```

```python
df_olidbr.to_csv("/content/drive/MyDrive/IC/data/processed/olid-br.csv", index=False)
```

# Hate-BR

```python
df_hatebr = pd.read_csv("/content/drive/MyDrive/IC/data/raw/hate-br.csv")
```

```python
df_hatebr.head()
```

```python
df_hatebr.drop(columns=['offensiveness_levels','offensive_&_non-hate_speech',
                        'non-offensive', 'specialist_1_hate_speech',
                        'specialist_2_hate_speech', 'specialist_3_hate_speech'], inplace=True)
```

```python
colunms_to_map = df_hatebr.columns[1:]
```

```python
df_hatebr[colunms_to_map] = df_hatebr[colunms_to_map].map(lambda x: int(x))
```

```python
df_hatebr.head()
```

```python
df_hatebr["instagram_comments"] = df_hatebr["instagram_comments"].apply(prep.process_tweet)
```

```python
df_hatebr.dropna(inplace=True)
```

```python
df_hatebr.to_csv("/content/drive/MyDrive/IC/data/processed/hate-br.csv", index=False)
```

# Interfaces
